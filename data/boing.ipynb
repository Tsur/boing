```python
import matplotlib.pyplot as plt
from numpy import genfromtxt

data=genfromtxt("jumplog1.csv", delimiter=',', names=['acc1','acc2', 'timestamp'])
x=data['timestamp']
y=data['acc2']
ax=plt.subplot(111)
ax.plot(x, y)
plt.show()
```
